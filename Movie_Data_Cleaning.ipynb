## Take this data and make a unique csv for each 'Type' of movie

```python
unique = df["Type"].unique()
len(unique)

for i in unique:
    dfi = df[df['Type'] == i]
    file_name = "Movie_Type_" + str(i) + "_.csv"
    dfi.to_csv(file_name, index = True)
    print(dfi)

```

## Now read all those files back into python and set the axis = 0 in pd.concat

```python
li = []
for i in unique: 
    file_name = "Movie_Type_" + str(i) + "_.csv"
    df = pd.read_csv(file_name)
    li.append(df)
df2= pd.concat(li, axis = 0)
print(df2)

```
